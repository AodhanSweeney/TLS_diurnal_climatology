```python
import numpy as np
import plotting
```

```python
# Load Colocated Data for each season
#####################################
# DJF
ERA_5_colocations_DJF = np.load('DJF_colocations_5x10_boxes.npy')
ERA_5_colocations_DJF_zm = np.nanmean(ERA_5_colocations_DJF, axis=1)
ERA_5_colocations_no_removal_DJF = np.load('DJF_colocations_5x10_no_removal.npy')
ERA_5_colocations_no_removal_DJF_zm = np.nanmean(ERA_5_colocations_no_removal_DJF, axis=1)
# MAM 
ERA_5_colocations_MAM = np.load('MAM_colocations_5x10_boxes.npy')
ERA_5_colocations_MAM_zm = np.nanmean(ERA_5_colocations_MAM, axis=1)
ERA_5_colocations_no_removal_MAM = np.load('MAM_colocations_5x10_no_removal.npy')
ERA_5_colocations_no_removal_MAM_zm = np.nanmean(ERA_5_colocations_no_removal_MAM, axis=1)
# JJA
ERA_5_colocations_JJA = np.load('JJA_colocations_5x10_boxes.npy')
ERA_5_colocations_JJA_zm = np.nanmean(ERA_5_colocations_JJA, axis=1)
ERA_5_colocations_no_removal_JJA = np.load('JJA_colocations_5x10_no_removal.npy')
ERA_5_colocations_no_removal_JJA_zm = np.nanmean(ERA_5_colocations_no_removal_JJA, axis=1)


# Load ERA-5 all Data
#####################################
ERA_5_alldata_DJF = np.load('DJF_ERA5_5x10_boxes_diurnal_cycles.npy')
ERA_5_alldata_DJF_zm = np.nanmean(ERA_5_alldata_DJF, axis=1)
ERA_5_alldata_MAM = np.load('MAM_ERA5_5x10_boxes_diurnal_cycles.npy')
ERA_5_alldata_MAM_zm = np.nanmean(ERA_5_alldata_MAM, axis=1)
ERA_5_alldata_JJA = np.load('JJA_ERA5_5x10_boxes_diurnal_cycles.npy')
ERA_5_alldata_JJA_zm = np.nanmean(ERA_5_alldata_JJA, axis=1)

# Load Radio Occultation Data
#####################################
gpsro_DJF = np.load('DJF_GPSRO_5_10_boxes_diurnal_cycles.npy')
gpsro_DJF_zm = np.nanmean(gpsro_DJF, axis=1)
gpsro_MAM = np.load('MAM_GPSRO_5_10_boxes_diurnal_cycles.npy')
gpsro_MAM_zm = np.nanmean(gpsro_MAM, axis=1)
gpsro_JJA = np.load('JJA_GPSRO_5_10_boxes_diurnal_cycles.npy')
gpsro_JJA_zm = np.nanmean(gpsro_JJA, axis=1)
gpsro_SON = np.load('SON_GPSRO_5_10_boxes_diurnal_cycles.npy')
gpsro_SON_zm = np.nanmean(gpsro_SON, axis=1)

# Load CCM3 Data
#####################################
ccsm_DJF = np.load('../ccsm_djf_TLS_5x10_boxes.npy')
ccsm_MAM = np.load('../ccsm_mam_TLS_5x10_boxes.npy')
ccsm_JJA = np.load('../ccsm_jja_TLS_5x10_boxes.npy')
ccsm_SON = np.load('../ccsm_son_TLS_5x10_boxes.npy')
ccsm_DJF_zm = np.nanmean(ccsm_DJF, axis=1)
ccsm_MAM_zm = np.nanmean(ccsm_MAM, axis=1)
ccsm_JJA_zm = np.nanmean(ccsm_JJA, axis=1)
ccsm_SON_zm = np.nanmean(ccsm_SON, axis=1)

# Load WACCM6 Data
#####################################
waccm6_DJF = np.load('../../../../WACCM_diurnal_cycle/DJF_WACCM6_TLS_diurnal_cycle.npy')
waccm6_MAM = np.load('../../../../WACCM_diurnal_cycle/MAM_WACCM6_TLS_diurnal_cycle.npy')
waccm6_JJA = np.load('../../../../WACCM_diurnal_cycle/JJA_WACCM6_TLS_diurnal_cycle.npy')
waccm6_SON = np.load('../../../../WACCM_diurnal_cycle/SON_WACCM6_TLS_diurnal_cycle.npy')
waccm6_DJF[-1] = waccm6_DJF[-2]
waccm6_MAM[-1] = waccm6_MAM[-2]
waccm6_JJA[-1] = waccm6_JJA[-2]
waccm6_SON[-1] = waccm6_SON[-2]
waccm6_DJF_zm = np.nanmean(waccm6_DJF, axis=1)
waccm6_MAM_zm = np.nanmean(waccm6_MAM, axis=1)
waccm6_JJA_zm = np.nanmean(waccm6_JJA, axis=1)
waccm6_SON_zm = np.nanmean(waccm6_SON, axis=1)
```

```python
import matplotlib.pyplot as plt

thinguy = gpsro_MAM[23:26,25:28,:]
thing = np.nanmean(np.reshape(thinguy, (len(thinguy)*len(thinguy[0]), 8)), axis=0)
plt.plot(thing)
```

```python
# Bin all seasons of data into 10 latitude bands
# DJF
ERA_5_cycles_DJF = plotting.area_bins(ERA_5_alldata_DJF_zm)
ERA_5_colocation_cycles_DJF = plotting.area_bins(ERA_5_colocations_DJF_zm)
ERA_5_colocations_no_removal_DJF = plotting.area_bins(ERA_5_colocations_no_removal_DJF_zm)
gpsro_cycles_DJF = plotting.area_bins(gpsro_DJF_zm)
ccsm_cycles_DJF = plotting.area_bins(ccsm_DJF_zm)
waccm6_cycles_DJF = plotting.area_bins(waccm6_DJF_zm)
# MAM
ERA_5_cycles_MAM = plotting.area_bins(ERA_5_alldata_MAM_zm)
ERA_5_colocation_cycles_MAM = plotting.area_bins(ERA_5_colocations_MAM_zm)
ERA_5_colocations_no_removal_MAM = plotting.area_bins(ERA_5_colocations_no_removal_MAM_zm)
gpsro_cycles_MAM = plotting.area_bins(gpsro_MAM_zm)
ccsm_cycles_MAM = plotting.area_bins(ccsm_MAM_zm)
waccm6_cycles_MAM = plotting.area_bins(waccm6_MAM_zm)
# JJA
ERA_5_cycles_JJA = plotting.area_bins(ERA_5_alldata_JJA_zm)
gpsro_cycles_JJA = plotting.area_bins(gpsro_JJA_zm)
ccsm_cycles_JJA = plotting.area_bins(ccsm_JJA_zm)
ERA_5_colocation_cycles_JJA = plotting.area_bins(ERA_5_colocations_JJA_zm)
ERA_5_colocations_no_removal_JJA = plotting.area_bins(ERA_5_colocations_no_removal_JJA_zm)
waccm6_cycles_JJA = plotting.area_bins(waccm6_JJA_zm)
# SON
#ERA_5_cycles_SON = plotting.area_bins(ERA_5_alldata_SON_zm)
gpsro_cycles_SON = plotting.area_bins(gpsro_SON_zm)
ccsm_cycles_SON = plotting.area_bins(ccsm_SON_zm)
#ERA_5_colocation_cycles_SON = plotting.area_bins(ERA_5_colocations_SON_zm)
#ERA_5_colocations_no_removal_SON = plotting.area_bins(ERA_5_colocations_no_removal_SON_zm)
waccm6_cycles_SON = plotting.area_bins(waccm6_SON_zm)
```
