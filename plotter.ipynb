```python
import numpy as np
import plotting
import matplotlib.pyplot as plt
```

```python
# Load Colocated Data for each season
#####################################
# DJF
ERA_5_colocations_DJF = np.load('DJFfiles/DJF_colocations_5x10_boxes.npy')
ERA_5_colocations_DJF_zm = np.nanmean(ERA_5_colocations_DJF, axis=1)
ERA_5_colocations_no_removal_DJF = np.load('DJFfiles/DJF_colocations_5x10_no_removal.npy')
ERA_5_colocations_no_removal_DJF_zm = np.nanmean(ERA_5_colocations_no_removal_DJF, axis=1)
# MAM 
ERA_5_colocations_MAM = np.load('MAMfiles/MAM_colocations_5x10_boxes.npy')
ERA_5_colocations_MAM_zm = np.nanmean(ERA_5_colocations_MAM, axis=1)
ERA_5_colocations_no_removal_MAM = np.load('MAMfiles/MAM_colocations_5x10_no_removal.npy')
ERA_5_colocations_no_removal_MAM_zm = np.nanmean(ERA_5_colocations_no_removal_MAM, axis=1)
# JJA
ERA_5_colocations_JJA = np.load('JJAfiles/JJA_colocations_5x10_boxes.npy')
ERA_5_colocations_JJA_zm = np.nanmean(ERA_5_colocations_JJA, axis=1)
ERA_5_colocations_no_removal_JJA = np.load('JJAfiles/JJA_colocations_5x10_no_removal.npy')
ERA_5_colocations_no_removal_JJA_zm = np.nanmean(ERA_5_colocations_no_removal_JJA, axis=1)
# SON
ERA_5_colocations_SON = np.load('SON_colocations_5x10_boxes.npy')
ERA_5_colocations_SON_zm = np.nanmean(ERA_5_colocations_SON, axis=1)
ERA_5_colocations_no_removal_SON = np.load('SON_colocations_5x10_no_removal.npy')
ERA_5_colocations_no_removal_SON_zm = np.nanmean(ERA_5_colocations_no_removal_SON, axis=1)


# Load ERA-5 all Data
#####################################
ERA_5_alldata_DJF = np.load('DJFfiles/DJF_ERA5_5x10_boxes_diurnal_cycles.npy')
ERA_5_alldata_DJF_zm = np.nanmean(ERA_5_alldata_DJF, axis=1)
ERA_5_alldata_MAM = np.load('MAMfiles/MAM_ERA5_5x10_boxes_diurnal_cycles.npy')
ERA_5_alldata_MAM_zm = np.nanmean(ERA_5_alldata_MAM, axis=1)
ERA_5_alldata_JJA = np.load('JJAfiles/JJA_ERA5_5x10_boxes_diurnal_cycles.npy')
ERA_5_alldata_JJA_zm = np.nanmean(ERA_5_alldata_JJA, axis=1)
ERA_5_alldata_SON = np.load('SONfiles/SON_ERA5_5x10_boxes_diurnal_cycles.npy')
ERA_5_alldata_SON_zm = np.nanmean(ERA_5_alldata_SON, axis=1)

# Load Radio Occultation Data
#####################################
gpsro_DJF = np.load('DJFfiles/DJF_GPSRO_5_10_boxes_diurnal_cycles.npy')
gpsro_DJF_zm = np.nanmean(gpsro_DJF, axis=1)
gpsro_MAM = np.load('MAMfiles/MAM_GPSRO_5_10_boxes_diurnal_cycles.npy')
gpsro_MAM_zm = np.nanmean(gpsro_MAM, axis=1)
gpsro_JJA = np.load('JJAfiles/JJA_GPSRO_5_10_boxes_diurnal_cycles.npy')
gpsro_JJA_zm = np.nanmean(gpsro_JJA, axis=1)
gpsro_SON = np.load('SONfiles/SON_GPSRO_5_10_boxes_diurnal_cycles.npy')
gpsro_SON_zm = np.nanmean(gpsro_SON, axis=1)

# Load CCM3 Data
#####################################
ccsm_DJF = np.load('../ccsm_djf_TLS_5x10_boxes.npy')
ccsm_MAM = np.load('../ccsm_mam_TLS_5x10_boxes.npy')
ccsm_JJA = np.load('../ccsm_jja_TLS_5x10_boxes.npy')
ccsm_SON = np.load('../ccsm_son_TLS_5x10_boxes.npy')
ccsm_DJF_zm = np.nanmean(ccsm_DJF, axis=1)
ccsm_MAM_zm = np.nanmean(ccsm_MAM, axis=1)
ccsm_JJA_zm = np.nanmean(ccsm_JJA, axis=1)
ccsm_SON_zm = np.nanmean(ccsm_SON, axis=1)

# Load WACCM6 Data
#####################################
waccm6_DJF = np.load('../../../../WACCM_diurnal_cycle/DJF_WACCM6_TLS_diurnal_cycle.npy')
waccm6_MAM = np.load('../../../../WACCM_diurnal_cycle/MAM_WACCM6_TLS_diurnal_cycle.npy')
waccm6_JJA = np.load('../../../../WACCM_diurnal_cycle/JJA_WACCM6_TLS_diurnal_cycle.npy')
waccm6_SON = np.load('../../../../WACCM_diurnal_cycle/SON_WACCM6_TLS_diurnal_cycle.npy')
waccm6_DJF[-1] = waccm6_DJF[-2]
waccm6_MAM[-1] = waccm6_MAM[-2]
waccm6_JJA[-1] = waccm6_JJA[-2]
waccm6_SON[-1] = waccm6_SON[-2]
waccm6_DJF_zm = np.nanmean(waccm6_DJF, axis=1)
waccm6_MAM_zm = np.nanmean(waccm6_MAM, axis=1)
waccm6_JJA_zm = np.nanmean(waccm6_JJA, axis=1)
waccm6_SON_zm = np.nanmean(waccm6_SON, axis=1)
```

```python
j2 = 1082.63e-6 
ae = 6378137 #km
i = np.deg2rad(72)
alt = 550000
a = ae + alt
G = 6.674e-11
Me = 5.972e24
w = np.sqrt((G*Me)/(a**3))

precession_rate = -(3/2)*j2*((ae/a)**2)*np.cos(i)*w
precession_rate_perday = precession_rate*(24*60*60)*180/np.pi

true_precession = 0.9856 - precession_rate_perday
print(true_precession)
```
